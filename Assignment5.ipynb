```python
from skimage.io import imread
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
img=imread("C:/Users/Devdutto/Desktop/band1.gif")
```

```python
plt.imshow(img)
```

```python
print(img)
```

```python
img.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=4)
```

```python
principalComponents = pca.fit_transform(img)
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['princi compo 1', 'princi compo 2', 'princi compo 3', 'princi compo 4'])
```

```python
print(principalDf)
```

```python
new_img= np.array(principalDf)
```

```python
type(new_img)
```

```python
plt.figure(figsize=(15,20))
plt.imshow(new_img)
```
